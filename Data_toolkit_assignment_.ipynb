Q.17 Explain the difference between apply() and map() in Pandas.

Answer - In Pandas, both apply() and map() are used to apply functions to data, but they have different use cases and operate at different levels of the DataFrame or Series. Here’s a detailed comparison of the two:

**apply()**

*Functionality*:
 The apply() function is used to apply a function along the axis of a DataFrame (rows or columns) or to a Series. It can take a function that operates on an entire row or column.

*Usage:*
For a DataFrame, you can specify the axis parameter:
axis=0: Apply the function to each column (default).
axis=1: Apply the function to each row.
For a Series, it applies the function to each element.


**map()**

*Functionality:*
The map() function is primarily used with Series to apply a function element-wise. It is designed for transforming or mapping values in a Series.

*Usage:*
map() is used on a Series and applies a function to each element. It can also accept a dictionary or a Series to map values.

Q.21  Why is NumPy’s array slicing faster than Python’s list slicing?

Answer - NumPy's array slicing is generally faster than Python's list slicing due to several key factors:

1.  Contiguous Memory Allocation: NumPy arrays are stored in contiguous blocks of memory, which allows for efficient access and manipulation of data. This contiguous storage enables faster iteration and slicing operations compared to Python lists, which can be composed of elements stored in different memory locations.

2.  Fixed Data Type: NumPy arrays have a fixed data type for all elements, which allows for optimized storage and computation. This uniformity enables NumPy to perform operations more efficiently, as it can leverage low-level optimizations and vectorized operations. In contrast, Python lists can contain elements of different types, which adds overhead for type checking and management.

3.  Vectorized Operations: NumPy is designed for numerical computations and supports vectorized operations, which means that operations can be applied to entire arrays at once without the need for explicit loops. This leads to significant performance improvements, especially for large datasets. Python lists, on the other hand, require explicit iteration for element-wise operations.

4.  Optimized C Implementation: NumPy is implemented in C, which allows it to take advantage of low-level optimizations and efficient memory management. The underlying C code can execute operations much faster than equivalent Python code, which is interpreted and has more overhead.

5.  Slicing Mechanism: When you slice a NumPy array, it creates a view of the original array rather than a copy. This means that slicing operations are performed without the overhead of copying data, making them faster. In contrast, Python list slicing creates a new list that contains the sliced elements, which involves copying the data and can be slower.

6.  Broadcasting: NumPy supports broadcasting, which allows for operations on arrays of different shapes without the need for explicit replication of data. This feature can lead to more efficient computations compared to Python lists, where such operations would require manual handling.

Q.4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

Answer -

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (using the data from the previous cell)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 30, 22, 35],
    'Salary': [50000, 60000, 55000, 70000]
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Compute correlation matrix, including only numeric columns
correlation_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10, 8))  # Set figure size
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", vmin=-1, vmax=1)

# Add title
plt.title('Correlation Heatmap', fontsize=16)

# Display heatmap
plt.show()
```

Q.5 Generate a bar plot using Plotly?

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Category': ['Electronics', 'Clothing', 'Home Goods', 'Books', 'Sports'],
    'Sales': [12500, 8700, 11200, 5600, 9800],
    'Profit': [4200, 3100, 3800, 1500, 2800]
}

df = pd.DataFrame(data)

# Create the bar plot
fig = px.bar(df,
             x='Category',
             y='Sales',
             color='Profit',
             title='Sales by Category with Profit Indicators',
             labels={'Sales': 'Total Sales ($)', 'Profit': 'Profit ($)'},
             text_auto='.2s',  # Auto-format text labels
             color_continuous_scale='Tealgrn',
             height=500)

# Customize layout
fig.update_layout(
    plot_bgcolor='white',
    xaxis_title="Product Category",
    yaxis_title="Total Sales ($)",
    xaxis={'categoryorder':'total descending'},
    hovermode='x unified',
    margin=dict(l=50, r=50, t=80, b=50),
    title_font=dict(size=20, family='Arial')
)

# Add interactive features
fig.update_traces(
    marker_line_color='darkgray',
    marker_line_width=1,
    hovertemplate="<b>%{x}</b><br>" +
                  "Sales: %{y:$,.0f}<br>" +
                  "Profit: %{customdata[0]:$,.0f}<extra></extra>",
    customdata=df[['Profit']]
)

# Show the plot
fig.show()
```

Q.7 Write a program to perform element-wise multiplication of two NumPy array.


```python
import numpy as np

def elementwise_multiplication(arr1, arr2):
    """
    Performs element-wise multiplication of two NumPy arrays.

    Parameters:
    arr1 (numpy.ndarray): First input array
    arr2 (numpy.ndarray): Second input array

    Returns:
    numpy.ndarray: Result of element-wise multiplication
    """
    # Check if arrays have the same shape
    if arr1.shape != arr2.shape:
        raise ValueError("Input arrays must have the same shape for element-wise multiplication")

    # Perform element-wise multiplication
    result = np.multiply(arr1, arr2)
    # Alternatively: result = arr1 * arr2

    return result

# Example usage
if __name__ == "__main__":
    # Create two sample arrays
    array1 = np.array([1, 2, 3, 4])
    array2 = np.array([5, 6, 7, 8])

    # Perform element-wise multiplication
    try:
        multiplied_array = elementwise_multiplication(array1, array2)
        print("Array 1:", array1)
        print("Array 2:", array2)
        print("Element-wise multiplication result:", multiplied_array)
    except ValueError as e:
        print("Error:", e)
```

Q.8  Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)  # X-axis values from 0 to 10
y1 = np.sin(x)               # First line: sine wave
y2 = np.cos(x)               # Second line: cosine wave
y3 = np.sin(x) * np.cos(x)   # Third line: product of sine and cosine

# Create the plot
plt.figure(figsize=(10, 6))  # Set figure size

# Plot each line with different styles and labels
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.plot(x, y3, label='sin(x)*cos(x)', color='green', linestyle=':', linewidth=2)

# Add title and labels
plt.title('Multiple Lines Plot Example', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Add legend
plt.legend(fontsize=10)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

 Q.10 Create a histogram using Seaborn to visualize a distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set style for better aesthetics
sns.set(style="whitegrid", palette="pastel")

# Generate sample data (normally distributed with some outliers)
np.random.seed(42)
data = np.concatenate([np.random.normal(loc=50, scale=10, size=500),
                       np.random.normal(loc=100, scale=5, size=50)])

# Create the histogram plot
plt.figure(figsize=(10, 6))

# Create histogram with density curve
hist_plot = sns.histplot(data=data, bins=30, kde=True,
                        edgecolor='white', linewidth=0.5,
                        color='skyblue', alpha=0.8)

# Add title and labels
plt.title('Distribution of Sample Data', fontsize=16, pad=20)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Customize the KDE curve
if hist_plot.lines:  # Check if KDE line exists
    hist_plot.lines[0].set_color('darkred')
    hist_plot.lines[0].set_linewidth(2)

# Add mean and median lines
mean_val = np.mean(data)
median_val = np.median(data)
plt.axvline(mean_val, color='green', linestyle='--', linewidth=2,
            label=f'Mean: {mean_val:.1f}')
plt.axvline(median_val, color='purple', linestyle=':', linewidth=2,
            label=f'Median: {median_val:.1f}')

# Add legend and adjust layout
plt.legend(fontsize=10)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd

try:
    # Load CSV with additional options
    df = pd.read_csv('your_file.csv',
                    encoding='utf-8',  # Specify encoding if needed
                    na_values=['NA', 'missing'])  # Handle missing values

    print(f"Successfully loaded data with {len(df)} rows and {len(df.columns)} columns")

    # Display first 5 rows with formatting
    with pd.option_context('display.max_columns', None):  # Show all columns
        print("\nFirst 5 rows:")
        print(df.head())

    # Display basic info
    print("\nDataFrame info:")
    print(df.info())

except FileNotFoundError:
    print("Error: File not found. Please check the file path.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

Q.13 Create a 3D scatter plot using Plotly.

```python
import plotly.graph_objects as go
import numpy as np

# Generate sample data
np.random.seed(42)
n_points = 100
x = np.random.randn(n_points)
y = np.random.randn(n_points)
z = np.random.randn(n_points)
colors = np.random.rand(n_points)
sizes = np.random.randint(5, 20, n_points)
categories = np.random.choice(['A', 'B', 'C'], n_points)

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=sizes,
        color=colors,                # Color by value
        colorscale='Viridis',        # Color scale
        opacity=0.8,
        colorbar=dict(title='Color Scale')
    ),
    text=[f'Point {i}' for i in range(n_points)],  # Hover text
    hoverinfo='text+x+y+z',
    customdata=categories,
    hovertemplate=
    "<b>%{text}</b><br><br>" +
    "X: %{x}<br>" +
    "Y: %{y}<br>" +
    "Z: %{z}<br>" +
    "Category: %{customdata}<extra></extra>"
)])

# Update layout for better visualization
fig.update_layout(
    title='3D Scatter Plot Example',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=0.1)  # Adjust camera view
        )
    ),
    width=800,
    height=600,
    margin=dict(r=20, l=10, b=10, t=30)
)

# Show the plot
fig.show()
```
